```python
def experience_agent(user_input, entered_experience, model="gpt-4.1-mini"):

    available_info = json.dumps(entered_experience)

    system_prompt = f"""
You extract work experience information ONLY from the current input: {user_input}

Each experience entry contains the following fields:
1. Company
2. Role
3. Start_Date
4. End_Date
5. Description


Your output must STRICTLY be JSON in this format:

{{
  "Company": ...,
  "Role": ...,
  "Start_Date": ...,
  "End_Date": ...,
  "Description": ...,
  "Next Question": <ask the next missing field OR ask if user has more experience>
}}

In case if any of the information is not avaliable then just fill it with "NA".


here are the available information {available_info} you have gathered so far,Identify which which fields are not available(NA).
your next question should be based on this field.

Rules:

1. Only fill fields that are currently "NA".
2. Never override fields already filled.
3. If user_input contains a company-like phrase (capitalized name, ends with X, Pvt, Ltd, Technologies, etc.) → treat as Company.
4. If user_input contains job title keywords (developer, analyst, engineer, manager, tester, intern) → treat as Role.
5. If user_input contains years or dates (2020, 2021, Jan, March, 01/2020) → treat as Start_Date or End_Date.
6. If user_input contains task/descriptions (“worked on”, “responsible for”, “handled”, “developed”) → treat as Description.
7. Fill only ONE missing field per turn.
8. Ask ONLY for the next missing field.
"""

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_input}
    ]

    response = openai.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0.0
    )

    content = response.choices[0].message.content

    try:
        data = json.loads(content)
    except:
        data = {}

    return data
```

```python
experiences = []

user_input = input("Do you have any work experience? (yes/no)\n")

if user_input.lower() == "no":
    print("You are marked as a Fresher.")
else:
    while True:

        current_exp = {
            "Company": "NA",
            "Role": "NA",
            "Start_Date": "NA",
            "End_Date": "NA",
            "Description": "NA"
        }

        user_input = input("Please provide your Company name:\n")

        for i in range(5):

            extract = experience_agent(user_input, current_exp)

            if current_exp["Company"] == "NA":
                current_exp["Company"] = extract["Company"]

            if current_exp["Role"] == "NA":
                current_exp["Role"] = extract["Role"]

            if current_exp["Start_Date"] == "NA":
                current_exp["Start_Date"] = extract["Start_Date"]

            if current_exp["End_Date"] == "NA":
                current_exp["End_Date"] = extract["End_Date"]

            if current_exp["Description"] == "NA":
                current_exp["Description"] = extract["Description"]

            user_input = input(extract["Next Question"] + "\n")

            if (current_exp["Company"] != "NA" and
                current_exp["Role"] != "NA" and
                current_exp["Start_Date"] != "NA" and
                current_exp["End_Date"] != "NA" and
                current_exp["Description"] != "NA"):

                experiences.append(current_exp)
                break

            if user_input == "quit":
                break

        more = input("Do you have more experience? (yes/no)\n")
        if more == "no":
            break
```
