```python
def create_about_me(subject, model = "gpt-4.1-nano"):
  #1. Create your prompt
  system_prompt ="""
  You are an expert resume builder and career branding specialist.
  Your task: craft a short, energetic, and professional "About Me" section for a resume targeting Data Science role

  Goal:
    Create a summary that grabs recruiters’ attention within seconds, reflects enthusiasm and professionalism,
    and showcases the candidate as a strong fit for the target Data Science or technology-related role.

    Summary should be of 3–4 sentences max
    Prioritize formatting clarity and readability.
    Ensure the resume looks polished, ATS-friendly, and visually balanced.
    Keep sentences short and powerful.
    No markdown or bullet points.

    """

  user_prompt = subject

  messages = [
      {"role": "system", "content": system_prompt},
      {"role": "user", "content": user_prompt}
  ]

  #3. Call OpenAI
  response = openai.chat.completions.create(model='gpt-4.1-nano', messages=messages)

  #4. Print the results
  response = response.choices[0].message.content
  display(Markdown(response))
  return(response)
```

```python
def master_resume_robot(about_me, projects, qualifications, model="gpt-4.1-nano"):
    system_prompt = """
   You are a professional resume compiler and formatter.

    Your task: Combine the provided sections — 'About Me', 'Projects', and 'Educational Qualifications' —
    into a cohesive, polished resume suitable for a Data Science or technology-related role.

    Produce a professional, consistent, and ATS-friendly resume that presents all sections clearly and flows naturally.

    """

    user_prompt = f"""
    Combine the following resume sections into a single professional resume.

    ABOUT ME:
    {about_me}

    PROJECTS:
    {projects}

    QUALIFICATIONS:
    {qualifications}
    """

    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]

    response = openai.chat.completions.create(model=model, messages=messages)
    final_resume = response.choices[0].message.content

    display(Markdown(final_resume))
    return final_resume
```

```python
import re
from fpdf import FPDF

class ResumePDF(FPDF):
    def footer(self):
        self.set_y(-15)
        self.set_font("Arial", "I", 8)
        self.cell(0, 10, f"Page {self.page_no()}", 0, 0, "C")

def export_resume_to_pdf(resume_text, filename="Final_Resume.pdf"):
    pdf = ResumePDF()
    pdf.add_page()

    pdf.set_auto_page_break(auto=True, margin=12)
    pdf.set_left_margin(15)
    pdf.set_right_margin(15)
    pdf.set_text_color(0, 0, 0)

    # Extract sections using regex (handles Skills or Qualifications merged)
    pattern = r"(ABOUT ME|PROJECTS|QUALIFICATIONS|SKILLS)\s*([\s\S]*?)(?=(ABOUT ME|PROJECTS|QUALIFICATIONS|SKILLS|$))"
    matches = re.findall(pattern, resume_text, re.IGNORECASE)

    for match in matches:
        section_title = match[0].upper().strip()
        section_content = match[1].strip()

        # Add section heading
        pdf.set_font("Arial", "B", 13)
        pdf.cell(0, 10, section_title, ln=True)
        pdf.ln(2)

        # Add section content
        pdf.set_font("Arial", size=11)
        pdf.multi_cell(0, 7, section_content)
        pdf.ln(4)

    # Save the PDF
    pdf.output(filename)
    print(f"✅ Resume saved as {filename}")
```

```python
final_resume = master_resume_robot(about_me, projects, qualifications)
export_resume_to_pdf(final_resume)
```
